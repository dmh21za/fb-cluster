<h3>0. Setup</h3>
<ul>
<li>Get the environment variables</li>
</ul>

```python
import os
from dotenv import load_dotenv
load_dotenv()

DATA_FILE_LOC = f"{os.getenv('DATA_DIR')}/{os.getenv('DATA_FILENAME')}"
```

<h3>1. Next we convert the CSV to a DF</h3>

```python
import pandas as pd

df = pd.read_csv(DATA_FILE_LOC)
# Let's just focus on photos because they generally have richer descriptions
df = df[df['Post type'] == 'Photos']
```

<h3>2. Now we choose what features we want to extract for now</h3>
<ul>
<li>Only > 1 distinct value</li>
<li>Only numeric values</li>
</ul>

```python
features = df.copy()

print(f"Initial length = {len(features.columns)}")
# Only keep columns with > 1 distinct value
features = features[[col for col in features.columns if features[col].nunique() > 1]]
print(f"Length distinct columns only = {len(features.columns)}")
# Only keep numeric values
features = features.select_dtypes(['number', 'bool'])
print(f"Length numeric-only = {len(features.columns)}")
# Get rid of some other columns e.g. aggregates and identifiers
exclude_columns = ['Post ID', 'Reactions, comments and shares']
features = features.drop(exclude_columns, axis=1)
```

<h3>4. Let's scale the features df</h3>

```python
from sklearn.preprocessing import RobustScaler
features = pd.DataFrame(
    RobustScaler().fit_transform(features),
    columns=features.columns,
    index=features.index
)
```
